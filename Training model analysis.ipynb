# Model Analysis  BERT-BASE-UNCASED

```python
# jupyter nbconvert yourNotebook.ipynb --no-input --no-prompt
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
metric = open('output/bert_base_metrics.pkl', 'rb')      
k = pickle.load(metric) 
metric.close()
```

## Training and Validation BCE-Loss Plot

```python
plt.plot(k['epochs'],k['train_loss'],c='red', label = 'training')
plt.plot(k['epochs'],k['val_loss'],c='blue', label = 'validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

#### As we can see, there is overfitting after the fifth epoch. At epoch 5, both training and validation are low. After that, validation loss is higher than training loss.

## Training and Validation F1-Micro-Score Plot

```python
plt.plot(k['epochs'],k['training_f1_micro'],c='red', label = 'training')
plt.plot(k['epochs'],k['val_f1_micro'],c='blue',label = 'validation')
plt.xlabel('Epochs')
plt.ylabel('F1-Micro-Score')
plt.title('Training and Validation F1-Micro-Score')
plt.legend()
plt.show()
```

#### Here also, we observe that the validation F1-Micro-Score after the fifth epoch does not increase while the training F1-Micro-Score does, indicating overfitting after the fifth epoch. The training F1-Micro-Score reached.90 at the tenth epoch.

## Training and Validation F1-Macro-Score Plot

```python
plt.plot(k['epochs'],k['training_f1_macro'],c='red', label = 'training')
plt.plot(k['epochs'],k['val_f1_macro'],c='blue',label = 'validation')
plt.xlabel('Epochs')
plt.ylabel('F1-Macro-Score')
plt.title('Training and Validation F1-Marco-Score')
plt.legend()
plt.show()
```

#### Here also, we can observe that training F1-Macro-Score increases after 5th epoch, indicating overfitting, but validation F1-Macro-Score does not increase after 5th epoch. Training F1-Macro-Score reached.85 at the 10th epoch.

## Training and Validation Hamming Loss Plot

```python
plt.plot(k['epochs'],k['training_hamming_loss'],c='red', label = 'training')
plt.plot(k['epochs'],k['val_hamming_loss'],c='blue',label = 'validation')
plt.xlabel('Epochs')
plt.ylabel('Hamming-Loss')
plt.title('Training and Validation Hamming-Loss')
plt.legend()
plt.show()
```

#### As we can see, there is overfitting after the fifth epoch. At epoch 5, both training and validation are low. After that, validation loss is higher than training loss.
